```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset_path = '../data/raw/hmnist_28_28_RGB.csv'
data = pd.read_csv(dataset_path)

print("First few rows of the dataset:")
print(data.head())
```

```python
class_distribution = data['label'].value_counts()
print("\nClass distribution:")
print(class_distribution)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=class_distribution.index, y=class_distribution.values)
plt.title('Distribution of Skin Lesion Classes')
plt.xlabel('Class Label')
plt.ylabel('Number of Images')
plt.show()
```

```python
pixel_columns = data.columns[:-1]  # Exclude the label column
pixel_values = data[pixel_columns].values.flatten()
print("\nRange of pixel values:")
print(f"Min: {pixel_values.min()}, Max: {pixel_values.max()}")
```

```python
def display_random_image(data, pixel_columns):
    random_index = np.random.randint(0, data.shape[0])
    image_data = data.iloc[random_index][pixel_columns].values
    image_data = image_data.reshape(28, 28, 3)
    plt.figure(figsize=(4, 4))
    plt.imshow(image_data.astype(np.uint8))
    plt.title(f"Class: {data.iloc[random_index]['label']}")
    plt.axis('off')
    plt.show()
    
print("\nDisplaying a random image from the dataset:")
display_random_image(data, pixel_columns)
```
